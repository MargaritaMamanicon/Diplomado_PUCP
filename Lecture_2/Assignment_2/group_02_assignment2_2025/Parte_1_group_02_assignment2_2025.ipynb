# 3. Importar dataset de Trabajo (500) y exploración de datos

```python
trabajo = pd.read_csv("./ENAHO-2023/Enaho01A-2023-500.csv", encoding="ISO-8859-10")
trabajo
```

```python
#Mostramos las primeras 5 filas
print (trabajo.head())
```

```python
#Convertimos en lista el nombre de las columnas
print(trabajo.columns.tolist())
```

```python
#Observamos el tipo de datos
print(trabajo.dtypes)
```

```python
#Seleccionamos una muestra
sub_trabajo=trabajo[['CONGLOME','VIVIENDA','HOGAR','CODPERSO','P501','P507','P513T']]
sub_trabajo

### P501 - La semana pasada, del ... al ..., ¿Tuvo Ud. algún trabajo? (Sin contar los quehaceres del hogar)
### P507 - Ud. se desempeño en su ocupación principal o negocio como:
### P513T - ¿Cuántas horas trabajó la semana pasada, en su ocupación principal, el día: Total
```

```python
#A. Cambiar el tipo de datos de una variable (por ejemplo, de texto a numérico).
print(sub_trabajo['P513T'].dtype)
```

```python
sub_trabajo['P513T'] = pd.to_numeric(sub_trabajo['P513T'], errors='coerce').fillna(0).astype(int)
```

```python
print(sub_trabajo['P513T'].dtype)
#Ahora la variable es numérica (tipo float)
```

```python
#B. Modificar algunos valores en una columna específica.
print(sub_trabajo['P507'].dtype)
print (sub_trabajo['P507'].head())
```

```python
sub_trabajo['P507'] = pd.to_numeric(sub_trabajo['P507'], errors='coerce').fillna(0).astype(int)
```

```python
sub_trabajo['P507'] = sub_trabajo['P507'].replace({
    0: "No responde / Missing",
    1: "Empleador o patrono",
    2: "Trabajador Independiente",
    3: "Empleado",
    4: "Obrero",
    5: "Trabajador Familiar No Remunerado",
    6: "Trabajador del Hogar",
    7: "Otro"
}).astype('category')
```

# 4. Fusión de Datasets (Merging Datasets)

```python
common_columns = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO']
```

```python
# Verificar registros coincidentes entre ambas submuestras
merged_keys = data[common_columns].merge(sub_trabajo[common_columns], on=common_columns, how='inner')
print(f"Registros coincidentes en claves comunes: {len(merged_keys)}")
```

```python
# Realizamos el merge
df_merged = pd.merge(data, sub_trabajo, on=common_columns, how='inner')

# Mostrar resultado
display(df_merged.head())
```

# 5. En el DataFrame resultante:

```python
#Agrupamos los datos por una variable utilizando groupby()
df_merged.groupby("P507")["P513T"].value_counts().reset_index()
```

```python
#Calcular un indicador estadístico relevante: Horas que trabajó en la semana según ocupación
resultado = df_merged.groupby('P507')['P513T'].mean().reset_index()

# Mostrar resultado
print("Promedio de horas de trabajo en la semana según ocupación:")
print(resultado)
```
