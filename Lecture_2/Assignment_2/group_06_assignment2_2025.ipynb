```python
# Select a subsample containing the variables ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO'] and between 3–5 additional variables of your interest.
variables = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO', 'P301A','P302','P306','P308B1']
subsample = data_300[variables].copy()
subsample
```

```python
# 2. Data Manipulation (Data Cleaning)
# Explore the DataFrame using summary functions (we use the DataFrame´s subsample)
subsample.describe()
```

```python
# Identify if there are missing values.
missings = subsample.isnull().sum().sum()
missings
```

```python
# Although there are no missing values at first glance, they appear with a specific codification
missing_codes = ["99", "9"] # En el diccionario, se indica que estos son los códigos de los mising values.
count_missing_codes = (subsample.isin(missing_codes)).sum().sum()
print("Cantidad total de missings:", count_missing_codes)
```

```python
# If they exist, remove them.
cleaned_sub = subsample.replace(missing_codes, np.nan)
cleaned_sub
```

```python
subsample_wo_na = cleaned_sub.dropna()
subsample_wo_na
```

```python
# 3. Import a second dataset (choose between Enaho01A-2023-200.csv or Enaho01A-2023-500.csv).
data_200 = pd.read_csv(r".\Enaho01-2023-200.csv",encoding="ISO-8859-10",low_memory=False)
```

```python
# Display the first 5 rows.
data_200.head()
```

```python
# Convert the column names into a list and print it.
list2_columns = data_200.columns.tolist()
print(list2_columns)
```

```python
# Check the data types.
print(data_200.dtypes)
```

```python
# Select a subsample containing the variables ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO'] and between 3–5 additional variables of your interest.
variables2 = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO', 'P207','P208A','P209','P210']
subsample2 = data_200[variables2].copy()
subsample2
```

```python
# Perform the following modifications:
# A. Change the data type of a variable (e.g., from text to numeric).
print(subsample2["P209"].dtype)
```

```python
subsample2["P209"] = pd.to_numeric(subsample2["P209"], errors="coerce")
```

```python
print(subsample2["P209"].dtype)
```

```python
# B. Modify some values in a specific column.
# In variable P207 (sex), we will modify the encoding to show numbers as strings: 1 == Hombre, 2 == Mujer. Modified: 1 == Mujer, 2 == Hombre.
subsample2["P207"] = pd.to_numeric(subsample2["P207"], errors="coerce")
subsample2["P207"] = subsample2["P207"].replace({
    1: "Hombre",
    2: "Mujer"
})
```

```python
print(subsample2["P207"])
```

```python
# Merging Datasets: [2 pts]
# Identify the common columns between the two datasets (from questions 16 and 18).
common_cols = subsample_wo_na.columns.intersection(subsample2.columns)
print(common_cols)
```

```python
# Verify whether the values match in both datasets. If not, correct the mismatched values to ensure a proper merge.
keys_300 = set(map(tuple, subsample_wo_na[common_cols].values))
keys_200 = set(map(tuple, subsample2[common_cols].values))

print("Keys de Módulo 300:", len(keys_300))
print("Keys de Módulo 200:", len(keys_200))
print("Intersección:", len(keys_300 & keys_200))
```

```python
# Recommendation: Use the following as common columns:
# common_columns = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO'] in pd.merge(..., on=common_columns, how=...).
# Perform the merge.
common_cols = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'CODPERSO']
merge = subsample_wo_na.merge(subsample2, on = common_cols, how = "inner")
```

```python
# Display the first 5 rows of the resulting DataFrame.
merge.head()
```

```python
# 5. In the resulting DataFrame:
# Group the data by a variable of your choice using groupby().
# Calculate a relevant statistical indicator, for example: the average income per category.

# Percentage of individuals who completed high school, by sex
merge["secundaria_completa"] = merge["P301A"] == 6
```

```python
tasa_secundaria_by_sex = merge.groupby("P207")["secundaria_completa"].mean()
tasa_secundaria_by_sex = tasa_secundaria_by_sex*100
print(tasa_secundaria_by_sex.round(1))
```
